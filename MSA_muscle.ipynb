```python
########## Alignment with MUSCLE ###########
###### Run with high memory, clush -> used alternatively MAFFT #######

import os
import glob
import subprocess

INPUT_DIR = "./msa_input_fastas"
OUTPUT_DIR = "./msa_aligned_results"
os.makedirs(OUTPUT_DIR, exist_ok=True)

def run_alignment():
    search_pattern = os.path.join(INPUT_DIR, "*_combined.fasta")
    fasta_files = glob.glob(search_pattern)
    
    if not fasta_files:
        print(f"No such file has been detected:'{INPUT_DIR}'.")
        return

    print(f"Total {len(fasta_files)} are aligning..")
    
    success_count = 0
    
    for input_fasta in fasta_files:
        base_name = os.path.basename(input_fasta).replace("_combined.fasta", "")
        output_aln = os.path.join(OUTPUT_DIR, f"{base_name}_aligned.fasta")
        
        # Set an alignment command in muscle
        cmd = ["muscle", "-align", input_fasta, "-output", output_aln]
        
        try:
            # Running alignment
            subprocess.run(cmd, capture_output=True, text=True, check=True)
            print(f"Alignment completed: {base_name}_aligned.fasta")
            success_count += 1
        except subprocess.CalledProcessError as e:
            # If there's an error
            print(f"Error occurs on ({base_name}):\n{e.stderr}")
            
    print(f"Alignment completed: (Total: {success_count}/{len(fasta_files)})")
    print(f"Saved in: {OUTPUT_DIR}")

if __name__ == "__main__":
    run_alignment()
```
